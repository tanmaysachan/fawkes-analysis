# Data Preproc

```python
df = pd.read_csv('football_data.csv')

df.keys()
```

```python
# Drop non numeric keys

keys_to_rem = [
   'Unnamed: 0','ID', 'Photo', 'Flag', 'Club', 'Club Logo', 
   'Special', 'Weak Foot', 'Work Rate', 'Body Type', 'Real Face',
   'Jersey Number', 'Joined', 'Loaned From', 'Contract Valid Until',
   'LS', 'ST', 'RS', 'LW', 'LF', 'CF', 'RF', 'RW',
   'LAM', 'CAM', 'RAM', 'LM', 'LCM', 'CM', 'RCM', 'RM', 'LWB', 'LDM',
   'CDM', 'RDM', 'RWB', 'LB', 'LCB', 'CB', 'RCB', 'RB', 'Release Clause'
]
```

```python
df = df.drop(keys_to_rem, axis=1).dropna().reset_index()
```

```python
len(df)
```

```python
Pos = []

# Sorted by position on the field.
# The lowest ones are closest to the goalkeeper, the highest ones are in the front to score.
for i in range(len(df)):
    if df['Position'][i] == 'LS' or df['Position'][i] == 'RS' \
    or df['Position'][i] == 'ST' or df['Position'][i] == 'CF':
        Pos.append(7)
    elif df['Position'][i] == 'LW' or df['Position'][i] == 'RW' \
    or df['Position'][i] == 'LF' or df['Position'][i] == 'RF' \
    or df['Position'][i] == 'LM' or df['Position'][i] == 'RM':
        Pos.append(6)
    elif df['Position'][i] == 'RAM' or df['Position'][i] == 'LAM' \
    or df['Position'][i] == 'CAM':
        Pos.append(5)
    elif df['Position'][i] == 'CM' or df['Position'][i] == 'LCM' \
    or df['Position'][i] == 'RCM':
        Pos.append(4)
    elif df['Position'][i] == 'RDM' or df['Position'][i] == 'LDM' \
    or df['Position'][i] == 'CDM':            
        Pos.append(3)
    elif df['Position'][i]=='LWB' or df['Position'][i]=='LB' \
    or df['Position'][i]=='RB' or df['Position'][i]=='RWB':
        Pos.append(2)
    elif df['Position'][i]=='RCB' or df['Position'][i]=='LCB' \
    or df['Position'][i]=='CB':
        Pos.append(1)
    else:
        Pos.append(0) # Goalkeeper

assert(len(Pos) == len(df))

df['Position'] = Pos

df['Foot'] = (df['Preferred Foot'] == 'Left').astype(int)
df = df.drop(['Preferred Foot'], axis=1)

def money_conv(x):
    if x[-1] == 'K':
        x = float(x[1:-1])*1000
    elif x[-1] == 'M':
        x = float(x[1:-1])*1000000
    else:
        x = 0
    return str(x)

df['Value'] = df['Value'].apply(money_conv)
df['Value'] = pd.to_numeric(df['Value'], errors='coerce')
        
df['Wage'] = df['Wage'].apply(money_conv)
df['Wage'] = pd.to_numeric(df['Wage'], errors='coerce')

def weight(x):
    return x.rstrip('lbs')

def height(x):
    return str(int(x[:1])*12 + int(x[2:]))

df['Weight'] = df['Weight'].apply(weight)
df['Weight'] = pd.to_numeric(df['Weight'], errors='coerce')

df['Height'] = df['Height'].apply(height)
df['Height'] = pd.to_numeric(df['Height'], errors='coerce')
    
```

# K-means

```python
kmean_samples = np.array(labeled_samples[list(numerical_unscaled.columns)])

def get_cluster(centres, idx):
    return np.sum((np.abs(kmean_samples - centres[idx]) ** 2), axis = 1) ** 0.5

def get_centre(clusters, idx):
    return np.mean(kmean_samples[np.argwhere(clusters == idx)], axis = 0)

def make_clusters(k, get_errors=False):
    
    # Randomly select centres from sampled random range
    centres = np.array([kmean_samples[i] for i in random.sample(range(len(kmean_samples)), k) ])
    
    prev = 10000000
    new = prev-1
    eps = 1e-7
    
    while prev - new > eps:
        if new != 9999999:
            centres = []
            for i in range(k):
                centres.append(get_centre(clusters, i))
            centres = np.array(centres)[:, -1, :]
            
        clusters = []
        prev = new
        new = 0

        for i in range(k):
            clusters.append(get_cluster(centres, i))
        clusters = np.argmin(np.array(clusters), axis = 0)
        
        new = np.sum(abs(kmean_samples - centres[clusters])**2, axis=1)**0.5
        new = new[0]
        
    if get_errors:
        return clusters, new
    else:
        return clusters
    
for k in [3, 5, 7]:
    print('\nClustering in', k, ' clusters')
    clusters = make_clusters(k)
    for c in range(1, k+1):
        subset = []
        cnt = 5
        for i in range(len(kmean_samples)):
            if clusters[i] == c-1 and cnt > 0:
                subset.append((labeled_samples['Name'][i], labeled_samples['Position'][i]))
                cnt -= 1
        print(subset)
```

(Based on previous football knowledge)
1. For the case of 3 clusters, we have players seperated in goalkeepers, strikers, defenders.
2. For the case of 5 clusters, we have strikers, keepers followed by weaker strikers, centre backs,  and weak wing players.
3. For the case of 7 clusters, we have some exceptional players in the first clusters implying a heavy rating based distinction. Cluster 2 and 4 seperate out the weak and the strong goalkeepers. The last cluster consists of good defenders whereas the cluster 5 tries to seperate out the weak defenders. Cluster 6 consists of weaker strikers.

```python
# Checking optimality till k = 4

clusters = []
for k in range(1, 7):
    c, e = make_clusters(k, get_errors=True)
    clusters.append((c, e))
```

## Divisive clustering


```python
def divisive_cluster(data):
    num = data.shape[0]*2-2
    data['cluster'] = num
    link = []
    sse = {}
    cur_data = data.drop('cluster', axis = 1)
    sse[num] = error(cur_data)
    to_splits = []
    for _ in range(data.shape[0]-1):
        to_split = max(sse, key=sse.get)
        
        split_index = (data.cluster==to_split)
        to_split_data = cur_data[split_index]
        
        
        parent_size = sum(split_index)
        kmeans = KMeans(n_clusters=2)
        kmeans = kmeans.fit(to_split_data)
        new_clusters = list(kmeans.predict(to_split_data))
        
        indexes_taken = [i for i, n in enumerate(split_index) if n]
        pos = [-1,-1]
        for ind , val in enumerate(new_clusters):
            pos[val] = indexes_taken[ind]
        if new_clusters.count(1)==1:
            replace_1 = pos[1]
        else:
            num -= 1
            replace_1 = num
        if new_clusters.count(0)==1:
            replace_0 = pos[0]                
        else:
            num -= 1
            replace_0 = num
        new_clusters = [replace_1 if j==1 else replace_0 for j in new_clusters]
        data.loc[split_index, 'cluster'] = new_clusters
        to_splits.append(to_split)        
        sse[replace_1] = error(cur_data[data.cluster==replace_1])
        sse[replace_0] = error(cur_data[data.cluster==replace_0])
                       
        link.append([replace_0,replace_1,sse[replace_0]+sse[replace_1],parent_size])
        sse.pop(to_split)
        
    link.reverse()
    to_splits.reverse()
    replacement_id = {}
    
    shape = data.shape[0]
    for i in range(shape):
        replacement_id[i] = i
        
    cnt = 0
    for n in to_splits:
        replacement_id[n] = cnt + shape
        to_splits[cnt] = cnt + shape
        cnt += 1
    new_link = np.array(link)
    
    cnt = 0
    for line in new_link:
        new_link[cnt][0] = replacement_id[line[0]]
        new_link[cnt][1] = replacement_id[line[1]]
        cnt += 1
        
    new_link[:,2] = np.sqrt(new_link[:,2])
    return to_splits, new_link

def error(x):
    return np.linalg.norm(np.array(x) - np.mean(np.array(x), axis = 0))
```

```python
sims = []

sim = cosine_similarity(samples)
for model in all_models:
    tot = 0
    cnt = 0
    for i in range(1000):
        for j in range(i):
            if model[i] == model[j]:
                tot += sim[i][j]
                cnt += 1
    sims.append(tot/cnt)
```

### Inter class similarity

```python
sims
```

```python
sims = []

for model in all_models:
    clusters = list(set([i for i in model if i != -1]))
    tot = 0
    cnt = 0
    intra = []
    for i1 in clusters:
        for i2 in clusters:
            if i1 != i2:
                m1 = np.mean(samples[(model == i1)], axis=0)
                m2 = np.mean(samples[(model == i2)], axis=0)
                
                similarity = np.dot(m1, m2)/(np.linalg.norm(m1)*np.linalg.norm(m2))
                tot += similarity
                cnt += 1
                intra.append(similarity)
    try:
        sims.append(tot/cnt)
    except:
        sims.append(0.999999999999)
```
